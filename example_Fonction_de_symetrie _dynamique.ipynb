```python
# Chargement des librairies
import pandas as pd
import matplotlib.pyplot as plt
import itertools as it
import numpy as np
```

```python
# Création d'un dataframe avec des données xdt et xgt qui sont les données de droite et de gauche à chaque instant t
xdt = np.random.randint(0,100,(100,))
xgt = np.random.randint(0,100,(100,))

#xgt = xdt # pour tester la symétrie vrai
#xgt = xdt + np.random.randint(0,2,(100,)) # test un légère symétrie aléatoire
#xgt = xdt + np.random.randint(0,100,(100,)) # test une grande symétrie aléatoire
df = pd.DataFrame({"xdt":xdt,"xgt":xgt})
```

```python
# Fonction de symétrie dynamique
def FSD(t):
  xdt = df.iloc[t,0] #df.iloc[ligne,colonne]
  xgt = df.iloc[t,1]
  rangexdt = max(df['xdt'])-min(df['xdt'])
  rangexgt = max(df['xgt'])-min(df['xgt'])
  return 2*(xdt-xgt)/(rangexdt+rangexgt)
```

```python
# Calcul de la fonction de symétrie dynamique et mise dans le dataframe : sym
sym = ['na'] * len(df)
for i in range(len(df)):
  sym[i] = FSD(i)
df['sym']=sym
```

```python
# Représentation graphique
plt.subplot(2,1,1)
plt.plot(df['xdt'],c='r')
plt.plot(df['xgt'],c='b')
plt.subplot(2,1,2)
plt.plot(df['sym'])
```

```python
# Fonction index de symétrie
def IS(t):
  xdt = df.iloc[t,0] #df.iloc[ligne,colonne]
  xgt = df.iloc[t,1]
  return 2*(xdt-xgt)/(xdt+xgt)

# Calcul de l'index de symétrie et mise dans le dataframe : sym2
sym2 = ['na'] * len(df)
for i in range(len(df)):
  sym2[i] = IS(i)
df['sym2']=sym2

# Représentation graphique
plt.subplot(2,1,1)
plt.plot(df['xdt'],c='r')
plt.plot(df['xgt'],c='b')
plt.subplot(2,1,2)
plt.plot(df['sym2'])
```

```python
# Comparaison fonction de symétrie dynamique et index de symétrie
plt.plot(df['sym'],c='r')
plt.plot(df['sym2'],c='b')
plt.legend(['Fonction de symétrie dynamique','Index de symétrie'])
```
